```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from os import listdir
from PIL import Image
from future.utils import iteritems
from scipy.stats import multivariate_normal as mvn
```

### Transform images into dataframe.

```python
def transform_into_df(path,images,width,height):
    labels=[]
    outdf=pd.DataFrame(columns=[*range(0,(width* height))])
    for image in images:
        img = Image.open(path+image)
        pixels = np.array(img.getdata())/255
        labels.append(ord(image[2])-ord('a'))
        outdf.loc[image]=pixels
    outdf['label']=labels
    return outdf
```

```python
train_df = transform_into_df('Problem_2_Dataset/Train/',listdir('Problem_2_Dataset/Train/'),12,12)
train_df.head(10)
```

```python
print("Train data has {} images with {} pixels/image.".format(train_df.shape[0],train_df.shape[1]-1) )
```

```python
test_df = transform_into_df('Problem_2_Dataset/Test/',listdir('Problem_2_Dataset/Test/'),12,12)
test_df.head()
```

```python
print("Test data has {} images with {} pixels/image.".format(test_df.shape[0],test_df.shape[1]-1) )
```

### Classifier's methods

```python
def fit(X, Y, add=1e-3):
    gauss = dict()
    out = dict()
    labels = set(Y)

    for c in labels:
        matched = X[Y == c]
        gauss[c] = {
            'mean': matched.mean(axis=0),
            'var': matched.var(axis=0) + add,
        }
        out[c] = float(len(Y[Y == c])) / len(Y)
    return gauss, out
```

```python
def predict(X,gauss,out): 
    K = len(gauss)
    prediction = np.zeros((X.shape[0], K))
    for c, g in iteritems(gauss):
        mean, var = g['mean'], g['var']
        prediction[:,c] = mvn.pdf(X, mean, np.sqrt(var)) * (out[c])
    return np.argmax(prediction, axis=1)
```

### Fit classifier

```python
Xtrain, Ytrain = (train_df[train_df.columns[:]]).values, (train_df['label']).values
Xtest, Ytest = (test_df[test_df.columns[:]]).values, (test_df['label']).values

gauss,out = fit(Xtrain, Ytrain)
```

### Predict classifier

```python
predictions = predict(Xtest,gauss,out)
```

### Score

```python
accuracy = np.mean(predictions == Ytest)*100
print("The accuracy is {0:.2f}%".format(accuracy))
```

### Plotting

```python
pred_list = list(predictions)
alphabet = []
y_axis = []
for char in range(0,26):
    alphabet.append(chr(ord('a')+char))
    y_axis.append(min(2,pred_list.count(char)))
```

```python
x_axis=np.arange(len(set(alphabet)))
plt.figure(num=None, figsize=(20, 7), dpi=80, facecolor='w', edgecolor='k')
plt.bar(alphabet, y_axis,alpha=0.8)
plt.yticks(range(0, 5))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Accuracy',fontsize=30)
plt.xlabel('Alphabet',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.savefig('Accuracy.jpg');
plt.show()
```
